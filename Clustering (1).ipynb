# Clustering (Unsupervised Machine Learning) 

Clustering is the task of dividing the population or data points into a number of groups such that data points in the same groups are more similar to other data points in the same group than those in other groups. In simple words, the aim is to segregate groups with similar traits and assign them into clusters.

Let’s understand this with an example. Suppose, you are the head of a rental store and wish to understand preferences of your costumers to scale up your business. Is it possible for you to look at details of each costumer and devise a unique business strategy for each one of them? Definitely not. But, what you can do is to cluster all of your costumers into say 10 groups based on their purchasing habits and use a separate strategy for costumers in each of these 10 groups. And this is what we call clustering.

# Overview
Online retail is a transactional data set, which contains all the transactions occurring between 01/12/2010 and 09/12/2011 for a UK-based and registered non-store online retail. The company mainly sells unique all-occasion gifts. Many customers of the company are wholesalers.

<h2>Business Goal</h2>
The aim to segement the Customers based on RFM ( Recency, Frequency, Monetary ) so that the company can target its customers efficiently

<h2>Attribute Information:</h2>

* **InvoiceNo:** Invoice number. Nominal, a 6-digit integral number uniquely assigned to each transaction. If this code starts with letter 'c', it indicates a cancellation.
* **StockCode:** Product (item) code. Nominal, a 5-digit integral number uniquely assigned to each distinct product.
* **Description:** Product (item) name. Nominal.
* **Quantity:** The quantities of each product (item) per transaction. Numeric.
* **InvoiceDate:** Invice Date and time. Numeric, the day and time when each transaction was generated.
* **UnitPrice:** Unit price. Numeric, Product price per unit in sterling.
* **CustomerID:** Customer number. Nominal, a 5-digit integral number uniquely assigned to each customer.
* **Country:** Country name. Nominal, the name of the country where each customer resides.

```python
# Larger ideas based on which we want to group customers

# High value customers -> Frequently and spend a lot in my store
# New customers - recently made a purchase and I want to hold on to them, marketing, discounts - permanent customers 
# Frequency, Recency and Amount
```

<h2>We are going to analysis the Customers based on below 3 factors:</h2>

* R (Recency): Number of days since last purchase
* F (Frequency): Number of tracsactions
* M (Monetary): Total amount of transactions (revenue contributed)

```python
# New Attribute : Monetary
retail['Amount'] = retail['Quantity']*retail['UnitPrice'] 
retail_m = retail.groupby('CustomerID')['Amount'].sum()
retail_m = retail_m.reset_index()
retail_m.head()
```

```python
# Number of Items they buy

# New Attribute : Frequency
retail_f = retail.groupby('CustomerID')['InvoiceNo'].count()
retail_f = retail_f.reset_index()
retail_f.columns = ['CustomerID', 'Frequency']
retail_f.head()
```

```python


# Number of Items they buy

# New Attribute : Frequency
retail_f = retail.groupby('CustomerID')['InvoiceNo'].count()
retail_f = retail_f.reset_index()
retail_f.columns = ['CustomerID', 'Frequency']
#retail_f.head()

# Merging the two dfs
retail_rfm = pd.merge(retail_m, retail_f, on='CustomerID', how='inner')
#retail_rfm.head()

# Data Cleaning - 
# 1. Amount 0 but 2 transactions - How is that possible ? -> reason
# 2. Amount > 0, but 0 transactions

retail_rfm = retail_rfm[~((retail_rfm['Frequency']!=0)&(retail_rfm['Amount']==0))]
retail_rfm = retail_rfm[~((retail_rfm['Amount']!=0)&(retail_rfm['Frequency']==0))]
retail_rfm.reset_index(drop = True, inplace = True)

#retail_rfm.head()

#retail.head(2)

# Convert to datetime to proper datatype
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'],format='%d-%m-%Y %H:%M')

# Compute the maximum date to know the last transaction date
max_date = max(retail['InvoiceDate'])
max_date
```

```python
# Compute the difference between max date and transaction date
retail['Diff'] = max_date - retail['InvoiceDate']
retail.head()

# Compute last transaction date to get the recency of customers
retail_p = retail.groupby('CustomerID')['Diff'].min()
retail_p = retail_p.reset_index()
retail_p.head()

# Extract number of days only
retail_p['Diff'] = retail_p['Diff'].dt.days
retail_p.head()

# Merge tha dataframes to get the final RFM dataframe
retail_rfm = pd.merge(retail_rfm, retail_p, on='CustomerID', how='inner')
retail_rfm.columns = ['CustomerID', 'Amount', 'Frequency', 'Recency']
retail_rfm.head()
```

# Model Building

## K-Means Clustering

K-means clustering is one of the simplest and popular unsupervised machine learning algorithms.<br>

The algorithm works as follows:

- First we initialize k points, called means, randomly.
- We categorize each item to its closest mean and we update the mean’s coordinates, which are the averages of the items categorized in that mean so far.
- We repeat the process for a given number of iterations and at the end, we have our clusters.

```python
# k-means with some arbitrary k - do you assume that your data has 5 clsuters , or 4 or 10 or just 1
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(df_scaled)
```

```python
kmeans.labels_
```

```python
df_scaled['Cluster'] = kmeans.labels_
```

```python
df_scaled
```

<h2>Finding the Optimal Number of Clusters</h2>

<h3>Elbow Curve to get the right number of Clusters</h3>

A fundamental step for any unsupervised algorithm is to determine the optimal number of clusters into which the data may be clustered. The Elbow Method is one of the most popular methods to determine this optimal value of k.

```python
# Elbow-curve/SSD
import warnings
warnings.filterwarnings('ignore')

ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(df_scaled)
    
    ssd.append(kmeans.inertia_)
    
# plot the SSDs for each n_clusters
ssd.insert(0,ssd[0])
plt.plot(ssd)
plt.show()
```

# Silhouette Analysis

$$\text{silhouette score}=\frac{p-q}{max(p,q)}$$

$p$ is the mean distance to the points in the nearest cluster that the data point is not a part of

$q$ is the mean intra-cluster distance to all the points in its own cluster.

* The value of the silhouette score range lies between -1 to 1. 

* A score closer to 1 indicates that the data point is very similar to other data points in the cluster, 

* A score closer to -1 indicates that the data point is not similar to the data points in its cluster.

```python
# 3 clusters (A, B, C) 
# -> I pick a point from Cluster A

#step_1
# from this point, I find out which is the nearest other cluster - B and C -> B
# I'm going to find out the average distance of this point to the points in the other cluster - B

#step-2
# -> I pick a point from Cluster A
# I find its average distance form all the points in it's own cluster
```

```python
# Silhouette analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(df_scaled)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(df_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
    
```

```python
# Final model with k=3
kmeans = KMeans(n_clusters=3, max_iter=50)
kmeans.fit(df_scaled)
```
